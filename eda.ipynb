```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the data
df = pd.read_csv('historic.csv')
```

```python
# Display basic information about the dataset
print(df.info())
print(df.describe())
```

```python
# Check for missing values
print(df.isnull().sum())
```

```python
# Analyze the distribution of the target variable
plt.figure(figsize=(8, 6))
df['success_indicator'].value_counts().plot(kind='bar')
plt.title('Distribution of Success Indicator')
plt.xlabel('Success')
plt.ylabel('Count')
plt.show()
```

```python
# Analyze the relationship between 'stars' and 'success_indicator'
plt.figure(figsize=(10, 6))
sns.boxplot(x='success_indicator', y='stars', data=df)
plt.title('Relationship between Stars and Success')
plt.show()
```

```python
# Analyze the distribution of products across categories
plt.figure(figsize=(12, 6))
df['category'].value_counts().plot(kind='bar')
plt.title('Distribution of Products Across Categories')
plt.xlabel('Category')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Analyze the relationship between 'main_promotion' and 'success_indicator'
plt.figure(figsize=(12, 6))
sns.countplot(x='main_promotion', hue='success_indicator', data=df)
plt.title('Success by Main Promotion')
plt.xlabel('Main Promotion')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
import numpy as np
```

```python
# Correlation heatmap for numeric features
numeric_df = df.select_dtypes(include = [np.number])
plt.figure(figsize=(10, 8))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Print unique values in categorical columns
categorical_columns = df.select_dtypes(include=['object']).columns
for col in categorical_columns:
    print(f"\nUnique values in {col}:")
    print(df[col].value_counts())
```
